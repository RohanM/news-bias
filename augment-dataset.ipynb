The smaller news dataset is more balanced, but is missing some centrist publications that the larger dataset covers. Augment the smaller dataset with some articles from these news sources.

```python
import csv
import pandas as pd
```

```python
batch_size = 100000
total_rows = 9000000
```

```python
news = pd.read_csv('data/all-the-news-2-1.csv', nrows=1000)
```

```python
news[news["publication"].isin(['Vox', 'Reuters'])]
```

```python
centrist_publications = ['Axios', 'CNBC', 'TechCrunch', 'The Hill', 'Wired']
```

```python
with open('data/articles4.csv', 'w') as csvfile:
    for i in range(0, total_rows, batch_size):
        print(i/1000000.0)
        news = pd.read_csv('data/all-the-news-2-1.csv', nrows=batch_size, skiprows=range(1, i))
        filtered_news = news[news["publication"].isin(centrist_publications)]
        filtered_news.to_csv(csvfile)
```
